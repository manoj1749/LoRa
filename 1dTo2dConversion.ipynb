```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
array = np.load('numpydata/sigfox_g0.0dB_att25dB_freq867.4MHz_4_slice94-N.npy')  
```

```python
plt.plot(array)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Plot of NumPy array')
plt.show()
```

```python
folder_path = 'data/'
for filename in os.listdir(folder_path):
    if filename.endswith('.npy'):
        array = np.load(os.path.join(folder_path, filename))
        real_part = np.real(array)
        imaginary_part = np.imag(array)
        array_2d = np.hstack((real_part[:, np.newaxis], imaginary_part[:, np.newaxis]))
        np.save(os.path.join(folder_path, filename), array_2d)
        print(f"Converted and saved {filename}")
print("Conversion completed.")
```
